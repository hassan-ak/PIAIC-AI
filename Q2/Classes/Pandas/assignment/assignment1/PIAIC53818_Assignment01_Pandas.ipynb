```python
# Loading all files
files = glob.glob("state*.csv")
files
```

```python
# making a single data frame from multiple files
us_census_list = []
for filename in files:
    data = pd.read_csv(filename)
    us_census_list.append(data)
us_census = pd.concat(us_census_list)
us_census
```

#### Regex to the Rescue

###### Step 05

```python
# Use regex to turn the Income column into a format that is ready for conversion into a numerical type.
```

```python
# view icome column
us_census.Income.head()
```

```python
# if $ signed is removed from the elements and converted into numeric form we can work with it for further steps
# so
```

```python
us_census.Income = us_census["Income"].replace("[\$,]","", regex = True)
us_census.Income.head()
```

```python
us_census.Income = pd.to_numeric(us_census.Income)
us_census.Income.head()
```

###### Step 06

```python
# Look at the GenderPop column. 
# We are going to want to separate this into two columns, the Men column, and the Women column.
# Split the column into those two new columns using str.split and separating out those results.
```

```python
us_census.GenderPop.head()
```

```python
us_census["str_split"] = us_census["GenderPop"].str.split("_")
us_census.head()
```

```python
us_census["Men"] = us_census.str_split.str.get(0)
us_census["Women"] = us_census.str_split.str.get(1)
us_census.head()
```

###### Step 07

```python
# Convert both of the columns into numerical datatypes.
# There is still an M or an F character in each entry! We should remove those before we convert.
```

```python
# first remove M and F from respective columns
us_census.Men = us_census["Men"].replace("[M]","", regex = True)
us_census.Women = us_census["Women"].replace("[F]","", regex = True)
us_census.head()
```

```python
us_census.dtypes
```

```python
# we need to convert Men and Women column to numeric form for the purpose of plotting the scatter plots
us_census.Men = pd.to_numeric(us_census.Men)
us_census.Women = pd.to_numeric(us_census.Women)
us_census.dtypes
```

```python
us_census.head()
```

```python
# Did you get an error? 
# These monstrous csv files probably have nan values in them! 
# Print out your column with the number of women per state to see.
# We can fill in those nans by using pandas’ .fillna() function.
# You have the TotalPop per state, and you have the Men per state. 
# As an estimate for the nan values in the Women column, 
# you could use the TotalPop of that state minus the Men for that state.
# Print out the Women column after filling the nan values to see if it worked!
```

```python
# we have a column of total Pop so we can use it to fill women column with the population of women
# Firsty we need to find if there is some Null Values in the total Pop column
```

```python
TotalPop_nan = us_census.TotalPop.isnull()
TotalPop_nan.value_counts()
```

```python
# As there is no Null values in the TotalPop column
# we can fill women population coulm using values of total pop and men pop
```

```python
us_census = us_census.fillna(value={"Women":us_census.TotalPop - us_census.Men})
```

```python
# now check if there is some null value in the column
fem_nan = us_census.Women.isnull()
fem_nan.value_counts()
```
